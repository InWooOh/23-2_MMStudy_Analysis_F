**Ch 4 딥러닝 시작**


**4-1 인공신경망 한계와 딥러닝 출현**

*   XOR 게이트의 경우 데이터가 비선형적으로 분리되기 때문에 분류가 어렵다
*   입력층과 출력층 사이에 하나 이상의 은닉층을 두어 비선형적으로 분리되는 데이터에 대해서도 학습이 가능하도록 다층 퍼셉트론을 사용



**4-2 딥러닝 구조**

**활성 함수**

*   시그모이드
  *   선형 함수 결과를 0~1 사이로 비선형 형태로 변형
  *   주로 로지스틱 회귀와 같은 분류 문제를 확률적으로 표현하는데 사용
  *   모델의 깊이가 깊어지면 기울 소멸 문제가 발생


*   하이퍼볼릭 탄젠트 함수
  *   선형 함수 결과를 -1~1 사이로 비선형 형태로 변형
  *   시그모이드에서 결과 값이 양수로 편항되는 문제 해결


*   렐루 함수(ReLU)
  *   입력(x)이 음수일 때는 0을 출력, 양수일 때는 x를 출력
  *   경사 하강법에 영향을 주지 않아 학습 속도가 빠르고 기울기 소멸 문제 해결

  
*   리키 렐루 함수(Leaky ReLU)함수
  *   입력 값이 음수이면 0이 아닌 0.001처럼 매우 작은 수 반환


*   소프트맥스 함수
  *   입력 값을 0~1 사이에 출력되도록 정규화 하여 출력 값들의 총합이 항상 1이 되도록

**은닉층이 많을 경우 딥러닝의 문제점**

*   과적합
  *   훈련 데이터에 대해 과하게 학습하여 실제 데이터에 대한 오차가 증가
  *   드롭아웃을 통해 일부 노드들을 학습에서 제외하여 해결


*   기울기 소멸 문제
  *   출력층에서 은닉층으로 전달되는 오차가 크게 줄어 학습이 되지 않는 현상
  *   렐루 활성화 함수를 사용하여 해결


*   성능 저하
  *   확률적 경사 하강법이나 미니 배치 경사하강법으로 해결



**4-3 딥러닝 알고리즘**

*   심층 신경망(DNN)
  *   입력층과 출력층 사이에 다수의 은닉층을 포함하는 인공 신경망

*   합성곱 신경망(CNN)
  *   영상 및 사진이 포함된 이미지 데이터에서 객체 탐지에 유용


*   순환 신경망(RNN)
  *   시계열 데이터(음악, 영상) 같은 시간 흐름에 따라 변화하는 데이터를 학습하는데 사용
  *   순환은 자기 자신은 참조한다는 것으로 현재 결과가 이전 결과와 연관 있다
